# **K-Means clustering in the IRIS dataset, with R.**

## **K-MEANS** : 
Utilizes a centroid point for clustering: **the mean of the group of points.**

<br>

It is rarely a real dataset point.

```
# activate R syntax
%load_ext rpy2.ipython
```

```
%%R
cluster = kmeans(iris[1:4], center=3)
cluster
```

```
%%R
confusion = table(iris$Species, cluster$cluster)
confusion
```

```
%%R
score = (confusion[1]+confusion[5]+confusion[9])/sum(confusion)
score
```

## And now we plot the default dispersion graph from R:


```
%%R
plot(iris[,1:4], col=cluster$cluster)
```
