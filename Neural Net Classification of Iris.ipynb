Changing the column names...

```r
names(myiris)[6] = 'setosa'
names(myiris)[7] = 'versicolor'
names(myiris)[8] = 'virginica'
```

Spliting the data into 70% for train and 30% for test.

```r
s = sample(2,150,replace=T,prob=c(0.7,0.3))
s
```

```r
myiristrain = myiris[s==1,]
myiristest = myiris[s==2,]
```

```r
dim(myiristrain)
```

```r
dim(myiristest)
```

Creating the model...

```r
model = neuralnet(setosa + versicolor + virginica ~ Sepal.Length + Sepal.Width + Petal.Length + Petal.Width, myiristrain, hidden=c(5,4))
```

```r
print(model)
```

```r
plot(neuralnet(setosa + versicolor + virginica ~ Sepal.Length + Sepal.Width + Petal.Length + Petal.Width, myiristrain, hidden=c(5,4)))
```

```r
plot(model, rep="best")
```

Now testing the model

```r
test = compute(model,myiristest[,1:4])
```

I don't want to pass the columns that I've used for classification, neither the species, so I choose all of the lines and columns from 1 to 4.

```r
test$net.result
```

It it visible that for each class it atributes a weight.

<br>

I'll create a dataframe with the result, and according to the column name, classificate the name of the corresponding column with the bigest attributed weight.

```r
result = as.data.frame(test$net.result)
```

```r
names(result)[1] = 'setosa'
names(result)[2] = 'versicolor'
names(result)[3] = 'virginica'
head(result)
```

Now I can create a class column, to compare success and failure, with the biggest weight, and then create a confusion matrix.

```r
result$class = colnames(result[,1:3])[max.col(result[,1:3],ties.method='first')]
```

```r
head(result)
```

```r
result
```

```r
confusion = table(result$class,myiristest$Species)
confusion
```

```r
sum(diag(confusion) * 100 / sum(confusion))
```
