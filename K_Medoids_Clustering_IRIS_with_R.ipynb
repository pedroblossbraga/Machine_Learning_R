## **K-Medoid** considers the most representative points of the dataset, to clusterize.

<br>

Unlike in K-Means (in which it its possible that the centroid does not belong to the dataset), in K-medoid, the medoid belongs to the dataset.

```
# activate R syntax
%load_ext rpy2.ipython
```

```
%%R
install.packages("cluster", dependencies=T)
library(cluster)
```

```
%%R
cluster = pam(iris[,1:4], k=3)
cluster
```

```
%%R
plot(cluster)
```

Or ploting separately:

```
%%R
clusplot(cluster)
```

```
%%R
plot(silhouette(cluster))
```

```
%%R
confusion = table(iris$Species, cluster$clustering)
confusion
```

```
%%R
score = (confusion[1]+confusion[5]+confusion[9])/sum(confusion)
score
```
